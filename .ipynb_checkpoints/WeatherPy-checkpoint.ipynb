```python

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from pprint import pprint
from scipy import stats

# Import API key
from api_keys import weather_api_key

weather_api_key = "your Key here"
g_key = "your Key here"

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#humidity
#clouds
#date
#max temp
#wind speed

# Save config information
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "Imperial"
counter = 1

#cities_df = pd.DataFrame(cities)
#Go over this with TA's to get the iterrows working
cities_df = pd.DataFrame()
cities_df["City"] = ""
cities_df["Cloudiness"] = ""
cities_df["Country"] = ""
#cities_df["Date"] = ""
cities_df["Humidity"] = ""
cities_df["Lat"] = ""
cities_df["Lng"] = ""
cities_df["Max Temp"] = ""
cities_df["Wind Speed"] = ""


#for index,counter in cities_df.iterrows():
#come back to this with TA.  


for i in cities:
    city = i
    query_url = f"{url}appid={weather_api_key}&q={city}&units={units}"
    weather_response = requests.get(query_url)
    weather_json = weather_response.json() #This is not calling the data every time.  If you add .json() and get on the same line as a variable that will call it.    
    #Print formatted better in json dumps
    #pprint(weather_json)
    try:
        cities_df.loc[city,'City'] = city
    except:
        cities_df.loc[city,'City']= None
    try:
        country = weather_json['sys']['country']
        cities_df.loc[city,'Country'] = country #the index is city defined here
    except:
        cities_df.loc[city,'Country'] = None
    try:
        clouds = weather_json['clouds']['all']
        cities_df.loc[city,'Cloudiness'] = clouds     
    except:
        cities_df.loc[city,'Cloudiness'] = None
    try:
        temp = weather_json['main']['temp_max']
        cities_df.loc[city,'Max Temp'] = temp
    except:
        cities_df.loc[city,'Max Temp'] = None
    try:
        wind = weather_json['wind']['speed']
        cities_df.loc[city,'Wind Speed'] = wind
    except:
        cities_df.loc[city,'Wind Speed'] = None
    try:
        humidity = weather_json['main']['humidity']
        cities_df.loc[city,'Humidity'] = humidity   
    except:
        cities_df.loc[city,'Humidity'] = None
    try:
        lat = weather_json['coord']['lat']
        cities_df.loc[city,'Lat'] = lat
    except:
        cities_df.loc[city,'Lat'] = None
    try:
        lng = weather_json['coord']['lon']
        cities_df.loc[city,'Lng'] = lng  
    except:
        cities_df.loc[city,'Lng'] = None
    print(f"The city is {city} and count is {counter} of total {len(cities)}")
    counter = counter + 1

cities_df.head()






#Method 2
# l_city = []
# l_clouds = []
# l_country = []
# l_date = []
# l_humidity = []
# l_lat = []
# l_lng = []
# l_temp = []
# l_wind = []
#print(cities_df)


```

#### Latitude vs. Temperature Plot

```python
temp = cities_df["Max Temp"].astype(float)
lat = cities_df["Lat"].astype(float)
lng = cities_df["Lng"].astype(float)
humidity = cities_df["Humidity"].astype(float)
cloudiness = cities_df["Cloudiness"].astype(float)
wind_speed = cities_df["Wind Speed"].astype(float)


#Temperature (f) vs Latitude
x_values = lat
y_values = temp
x_median = x_values.median()
y_median = y_values.median()
plt.scatter(x_values,y_values)
plt.title('All Temp vs Latitude')
plt.xlabel('Latitude')
plt.ylabel('Temperature (f)')
#Perform linear regression
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(x_values, y_values)
# Get regression values
# y = mx + b
regress_values = x_values * slope + intercept
#print(regress_values)
# Create line equation string
line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
print(line_eq)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(x_median,y_median),fontsize=15,color="red")
# Print r square value
print(f"The r-squared is: {round(rvalue,2)}")
# Show plot
plt.savefig("All Temp.png")
plt.show()
print('This is a scatter plot showing the relationship between temperature and latitude')
print('The warmest part of the earth is just north of the equator, latitude ~20')
print('Below the equator tempeartures vary between 40 and 80 degrees farenheight')
print('Above the equator temperatures varry between 0 and 100 faranehight for this sample set')
print(' ')
print(' ')


#Northern Hempisphere Temperature (f) vs Latitude)
temp_north = cities_north_df["Max Temp"].astype(float)
lat_north = cities_north_df["Lat"].astype(float)
x_values = lat_north
y_values = temp_north
x_median = x_values.median()
y_median = y_values.median()
plt.scatter(x_values,y_values)
plt.title('Northern Hemisphere Temperature vs Latitude')
plt.xlabel('Latitude')
plt.ylabel('Temperature (f)')
#Perform linear regression
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(x_values, y_values)
# Get regression values
# y = mx + b
regress_values = x_values * slope + intercept
#print(regress_values)
# Create line equation string
line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
print(line_eq)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(x_median,y_median),fontsize=15,color="red")
# Print r square value
print(f"The r-squared is: {round(rvalue,2)}")
# Show plot
plt.savefig("North Temp.png")
plt.show()


#Southern Hempisphere Temperature (f) vs Latitude)
temp_south = cities_south_df["Max Temp"].astype(float)
lat_south = cities_south_df["Lat"].astype(float)
x_values = lat_south
y_values = temp_south
x_median = x_values.median()
y_median = y_values.median()
plt.scatter(x_values,y_values)
plt.title('Southern Hemisphere Temperature vs Latitude')
plt.xlabel('Latitude')
plt.ylabel('Temperature (f)')
#Perform linear regression
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(x_values, y_values)
# Get regression values
# y = mx + b
regress_values = x_values * slope + intercept
#print(regress_values)
# Create line equation string
line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
print(line_eq)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(x_median,y_median),fontsize=15,color="red")
# Print r square value
print(f"The r-squared is: {round(rvalue,2)}")
# Show plot
plt.savefig("South Temp.png")
plt.show()
```
